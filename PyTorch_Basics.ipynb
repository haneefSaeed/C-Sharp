## SQUEEZE

You have this:
`[1, 2, 3]`        → shape `[3]`

`unsqueeze(0)` wraps it inside another list:
`[[1, 2, 3]]`      → shape `[1, 3]`

`unsqueeze(1)` makes each element its own box:
`[[1], [2], [3]]`  → shape `[3, 1]`

`squeeze()` removes all “pointless” dimension boxes:
`[[1, 2, 3]]`  → squeeze → `[1, 2, 3]`

```python
lst = torch.tensor([1,2,3])
lst, lst.shape
```

```python
lst = torch.unsqueeze(lst, 0)
```

```python
lst, lst.shape
```

```python
lst = torch.squeeze(lst, )
```

```python
lst, lst.shape
```

## PERMUTE
Used mostly in images

Example:
Drawer[Type][Color][Size]
Example shape:
[3 types, 4 colors, 2 sizes] → shape (3, 4, 2)
If you call: drawer.permute(1, 0, 2), It means:
now organize as:
Drawer[Color][Type][Size]




```python
# example

img = torch.rand(3, 4, 5)
img, img.shape
```

```python
img_hwcn = img.permute(1,2,0) ## index 2 => 5, 1=>4 , 0 => 3, we can rearrange
img_hwcn, img_hwcn.shape
```
